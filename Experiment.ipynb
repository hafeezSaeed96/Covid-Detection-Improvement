```python
import seaborn as sn
import matplotlib.pyplot as plt
import pandas as pd

def plotAccuracyPPV(covid, pneu, normal):

    plt.figure()
    plt.plot(covid, "g", label = "Covid-19")
    plt.plot(pneu, "r--", label = "Pneumonia")
    plt.plot(normal, "b:", label = "Normal")
    plt.xlabel("Epoch")
    plt.ylabel("PPV")
    plt.title("Positive Predictive Value for each Epoch")
    plt.legend()
    
    
def plotAccuracySENS(covid, pneu, normal):

    plt.figure()
    plt.plot(covid, "g", label = "Covid-19")
    plt.plot(pneu, "r--", label = "Pneumonia")
    plt.plot(normal, "b:", label = "Normal")
    plt.xlabel("Epoch")
    plt.ylabel("Sensitivity")
    plt.title("Sensitivity for each Epoch")
    plt.legend()
    
    

def plotLoss(train_Cost):
    
    plt.figure()
    plt.plot(train_Cost, "g", label = "Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Loss for each Epoch")
    plt.legend()
    plt.show()
    
    
def confusion(array):

    df_cm = pd.DataFrame(array, ["Normal", "Pneumonia", "Covid-19"], ["Normal", "Pneumonia", "Covid-19"])
    plt.figure(figsize=(8,5))
    df_cm.index.name = 'Predicted Values'
    df_cm.columns.name = 'Actual Values'
    sn.set(font_scale=1.4) # for label size
    sn.heatmap(df_cm, annot=True, annot_kws={"size": 16}) # font size

    plt.show()
    



```
